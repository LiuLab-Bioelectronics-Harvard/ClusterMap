# ClusterMap for STARmap human heart

```python
from ClusterMap.clustermap import *
from anndata import AnnData
import tifffile
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import scipy.io
```

### Read spots and dapi image

```python
### set file folder
filepath='./datasets/human_heart/'

### read dapi: col, row, z
dapi = tifffile.imread(filepath+'dapi.tif')
```

```python
dapi.shape
```

```python
### read spots
spots = pd.read_csv(filepath+'reads.csv')
spots.columns=['index','spot_location_1','spot_location_2','spot_location_3','gene_name']
```

```python
### convert gene_name to gene identity
genes=pd.DataFrame(spots['gene_name'].unique())
a1=list(genes[0])
gene=list(map(lambda x: a1.index(x)+1, spots['gene_name']))
spots['gene']=gene
spots['gene']=spots['gene'].astype('int')

# ### save gene annotation as genelist.csv
genes.to_csv(filepath+'genelist.csv', header=False, index=False)
```

```python
plt.figure(figsize=(6,6))
length=200
plt.imshow(np.sum(dapi,axis=2)[:length,:length],cmap='Greys')
test1=[i[1] and i[0] for i in zip(spots['spot_location_1']<length,spots['spot_location_2']<length)]

plt.scatter(spots['spot_location_1'][test1],spots['spot_location_2'][test1],s=0.5)
```

```python
plt.plot(dapi[130,125,:])
```

```python
### set radius parameters
xy_radius=35
z_radius=5
```

### Instantiate model

```python
### instantiate model
num_gene=np.max(spots['gene'])
gene_list=np.arange(1,num_gene+1)
num_dims=len(dapi.shape)
model = ClusterMap(spots=spots, dapi=dapi, gene_list=gene_list, num_dims=num_dims,
                   xy_radius=xy_radius,z_radius=z_radius,fast_preprocess=False)
```

### Show spatial distribution of marker genes

```python
###select gene markers
marker_genes=['MYL7','MYL4','VWF']
genes_list=list(genes[0])
```

```python
model.plot_gene(marker_genes,genes_list,figsize=(4,4),s=0.6)
```

### Pre-processing

```python
### pct_filter: percent of spots filtered out
### default: filter 10% spots out
pct_filter=0.1
model.preprocess(pct_filter)
```

```python
### plot preprocessing results
model.spots['is_noise']=model.spots['is_noise']+1
model.plot_segmentation(figsize=(10,10),s=0.6,method='is_noise',
                        cmap=np.array(((0,1,0),(1,0,0))),
                        plot_dapi=True)
model.spots['is_noise']=model.spots['is_noise']-min(model.spots['is_noise'])-1
```

### Cell segmentation

```python
%%time

### segmentation
model.min_spot_per_cell=5
model.segmentation(cell_num_threshold=0.003,dapi_grid_interval=3,add_dapi=True,use_genedis=True)

# res_over_dapi_erosion(self.spots, self.dapi_binary)
```

```python
### plot cell segmentation results in spots (colored by cells)
model.plot_segmentation(figsize=(10,10),s=1,plot_with_dapi=False,plot_dapi=True, show=False)
plt.show()
```

```python
model.create_convex_hulls(figscale=10,plot_with_dapi=True)
```

```python
###plot cells in 3D
model.plot_segmentation_3D(figsize=(8,8),elev=45, azim=-65)
```

```python
path_save = filepath+'spots_all.csv'
model.save_segmentation(path_save)
```

### Perform cell typing (based on Scanpy package)

```python
path_save = filepath+'spots_all.csv'
savespots=pd.read_csv(path_save)
model.spots=savespots
model.cellcenter = np.genfromtxt(filepath+"cellcenter.csv", delimiter=',')
```

```python
cellid='clustermap'
geneid='gene'
num_gene=np.max(spots['gene'])
gene_list=np.arange(1,num_gene+1)
num_dims=len(dapi.shape)

cluster_method='leiden'
```

```python
model.create_cell_adata(cellid,geneid,gene_list,genes,num_dims)
```

```python
model.cell_preprocess()
```

```python
model.cell_typing(cluster_method=cluster_method,resol=0.8)
```

```python
cluster_pl=model.plot_cell_typing()
```

```python
model.map_cell_type_to_spots(cellid)
```

```python
###plot cells in 3D
cluster_pl = sns.color_palette('tab20',len(model.cell_adata.obs['cell_type'].unique()))
model.plot_segmentation_3D(figsize=(8,8),elev=45, azim=-65,method='cell_type',
                          cmap=np.array(cluster_pl))
```
